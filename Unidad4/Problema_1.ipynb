```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Datos proporcionados
media = 70
desviacion_estandar = 4
n_estudiantes = 500

# Limites del intervalo de interés
x1 = 68
x2 = 72
```

```python
# Calcular la probabilidad con la función de distribución acumulada (CDF)
p1 = norm.cdf(x2, loc=media, scale=desviacion_estandar)
p2 = norm.cdf(x1, loc=media, scale=desviacion_estandar)

probabilidad = p1 - p2
print(f"Probabilidad de que un estudiante pese entre 68 y 72 kg: {probabilidad:.4f}")
```

```python
# Número esperado de estudiantes en ese rango
n_en_rango = probabilidad * n_estudiantes
print(f"Número estimado de estudiantes entre 68 y 72 kg: {n_en_rango:.0f}")
```

```python
# Crear valores para el eje x
x = np.linspace(media - 4*desviacion_estandar, media + 4*desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal', color='blue')
plt.fill_between(x, y, where=(x >= x1) & (x <= x2), color='skyblue', alpha=0.5, label='Área entre 68 y 72 kg')
plt.axvline(x1, color='green', linestyle='--', label='68 kg')
plt.axvline(x2, color='red', linestyle='--', label='72 kg')

# Mostrar la probabilidad en la gráfica
plt.text(media, max(y)*0.4, f'Probabilidad ≈ {probabilidad:.4f}', fontsize=12, ha='center', bbox=dict(facecolor='white', alpha=0.7))

plt.title('Distribución Normal del Peso de Estudiantes del ITD')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
