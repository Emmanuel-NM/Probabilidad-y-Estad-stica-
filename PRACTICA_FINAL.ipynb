```python
import pandas as pd

# Cargar grupo1.csv despus de haberlo subido
df1 = pd.read_csv("grupo1.csv")

# Ver las primeras filas
df1.head()
```

```python
# Estadsticas descriptivas del Grupo 1
df1.describe()
```

```python
# Verificar tipos de datos y valores nulos
df1.info()
```

```python
# Mostrar los nombres exactos de las columnas
df1.columns
```

```python
import matplotlib.pyplot as plt

# Histograma de la columna Calificacion
plt.hist(df1['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Calificaciones - Grupo 1')
plt.xlabel('Calificacin')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Preparar las variables
X = df1[['HorasEstudio']]  # variable independiente
y = df1['Calificacion']    # variable dependiente

# Crear el modelo y entrenarlo
modelo = LinearRegression()
modelo.fit(X, y)

# Obtener pendiente (coeficiente) e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print(f"Ecuacin de la recta: Calificacin = {pendiente:.2f} * HorasEstudio + {intercepto:.2f}")
```

```python
from scipy.stats import pearsonr

# Calcular el coeficiente de correlacin de Pearson
r, p_value = pearsonr(df1['HorasEstudio'], df1['Calificacion'])

print(f"Coeficiente de correlacin de Pearson: {r:.4f}")
print(f"Valor p: {p_value:.4f}")

```

```python
import matplotlib.pyplot as plt

# Grfico de dispersin
plt.scatter(df1['HorasEstudio'], df1['Calificacion'], color='blue', label='Datos reales')

# Lnea de regresin
plt.plot(df1['HorasEstudio'], modelo.predict(X), color='red', label='Lnea de regresin')

plt.title('Regresin Lineal - Grupo 1')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin')
plt.legend()
plt.grid(True)
plt.show()
```

####Cul es la interpretacin del coeficiente de regresin?
El coeficiente de regresin (pendiente) indica cunto se espera que aumente la calificacin final por cada hora adicional de estudio.
Por ejemplo, si la pendiente fue 1.5, entonces por cada hora de estudio extra, la calificacin aumenta en promedio 1.5 puntos.

2. Cmo se interpreta el coeficiente de correlacin?
El coeficiente de correlacin de Pearson (por ejemplo 0.85) indica una fuerte relacin positiva entre las horas de estudio y las calificaciones.
Si el valor estuviera ms cerca de 0, significara poca o ninguna relacin; si estuviera cerca de -1, sera una relacin negativa.

3. El modelo de regresin es estadsticamente significativo?
Si el valor p es menor que 0.05, se considera que la relacin entre las variables es estadsticamente significativa, es decir, no es producto del azar.

4. Qu tan confiable es el modelo para predecir?
El modelo parece confiable si la relacin es fuerte (correlacin alta) y el grfico muestra que los puntos siguen de cerca la lnea.
Sin embargo, como es un modelo simple, puede no ser preciso para casos extremos o para predecir ms all de los valores observados.
